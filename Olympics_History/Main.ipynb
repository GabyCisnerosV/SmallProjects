```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
events=pd.read_csv("Data\Olympic_Athlete_Event_Results.csv")

#Add Gender
temp_events=events["event"].str.split(",",expand=True)
temp_events["gender"]=""
for i in range(1,len(temp_events.columns)-1):
    temp_events["gender"]=np.where(temp_events[i].isna(),temp_events["gender"],temp_events[i])
events=pd.concat([events,temp_events[["gender"]]],axis=1)

events_grouped=events.groupby(["edition","sport","gender"]).agg({"event":"nunique"}).reset_index()
```

```python
num_cols=3
num_rows=int(np.ceil(events_grouped["sport"].nunique()/num_cols))

fig, axs=plt.subplots(figsize=(num_cols*5,num_rows*4),ncols=num_cols,nrows=num_rows)
x=0
y=0
for sp in events_grouped["sport"].unique():
    df_sp=events_grouped[events_grouped["sport"]==sp].reset_index()
    for g in df_sp["gender"].unique():
        axs[x,y].plot(df_sp[df_sp["gender"]==g]["edition"],df_sp[df_sp["gender"]==g]["event"],label=g)
    axs[x,y].set_title(sp)
    axs[x,y].set_xlabel("edition")
    axs[x,y].set_xlabel("Number of sports")
    axs[x,y].legend()
    plt.draw()
    axs[x,y].set_xticklabels(axs[x,y].get_xticklabels(), rotation=30, ha='right') 

    y+=1
    if y>num_cols-1:
        y=0
        x+=1
plt.tight_layout(pad=1.5)
plt.show()

```
